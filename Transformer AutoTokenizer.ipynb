```python
# Import AutoTokenixer form transformers
from transformers import AutoTokenizer
```

1. **Tokenizer**: A tokenizer is a tool used to break down text into smaller units, typically words or subwords. It is an essential component in NLP pipelines for tasks such as tokenization, where the input text is segmented into tokens that serve as the basic units of meaning.

2. **Pre-trained Model**: A pre-trained model is a model that has been trained on a large corpus of data for a specific NLP task, such as text classification, named entity recognition, or language generation. These models have learned representations of language that can be fine-tuned or used directly for downstream tasks.


```python
"""
Tokenizer and Pretrained model
"""
# name the preexisting tokenizer you wan to use
tokenizer_name = "bert-base-uncased"
#Use the from_pretrained method to load the tokenizer associated with a specific pre-trained model or tokenizer name
tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

text = "This is an example sentence."

# use .encode to encode a text
tokens = tokenizer.tokenize(text, add_special_tokens=True)
print(tokens)  
```


3. **Tokenizer Vocabulary**: The tokenizer vocabulary is the set of tokens used by the tokenizer to represent words or subwords in the input text. This vocabulary is usually derived from the corpus used to train the tokenizer or the pre-trained model.


```python
"""
tokenizer vocab
"""
#tokenizer.get_vocab()
tokenizer.vocab
```


4. **Token Number (ID)**: Each token in the tokenizer vocabulary is assigned a unique identifier or token number. This number serves as a way to represent the token internally within the NLP model. Token IDs are used during encoding and decoding processes.

```python
"""
geting id of a word in vocab
"""
tokenizer.vocab['deployed']
tokenizer.get_vocab().get("deployed")
```

5. **Encoding**: Encoding refers to the process of converting input text into a numerical representation that can be understood by the NLP model. This typically involves mapping each token in the input text to its corresponding token number (ID) in the tokenizer vorained models.

```python
"""
Encoiding
"""
text = "This is an example sentence."

# use .encode to encode a text
encoded_sequence = tokenizer.encode(text, add_special_tokens=True)
print(encoded_sequence)  
```

6. **Decoding**: Decoding is the reverse process of encoding, where the numerical representation generated by the model is converted back into human-readable text. This process involves mapping token IDs back to their corresponding tokens in the tokenizer vocabulary and reconstructing the orrained models.

```python
"""
Decode 
"""
print(tokenizer.decode(encoded_sequence))
```
