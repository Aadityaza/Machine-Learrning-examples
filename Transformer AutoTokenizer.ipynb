```python
# Import AutoTokenixer form transformers
from transformers import AutoTokenizer
```

```python
"""
tokenizer

pretrained model
"""
# name the tokenizer you wan to use
tokenizer_name = "bert-base-uncased"
#Use the from_pretrained method to load the tokenizer associated with a specific pre-trained model or tokenizer name
tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
```

```python
"""
tokenizer object
"""
tokenizer
```

```python
"""
tokenizer vocab
"""
#tokenizer.get_vocab()
tokenizer.vocab
```

```python
"""
geting id of a word in vocab
"""
tokenizer.vocab['deployed']
tokenizer.get_vocab().get("deployed")
```

```python
"""
tokenize
"""
text = "This is an example sentence."

# use .encode to encode a text
encoded_sequence = tokenizer.tokenize(text, add_special_tokens=True)
print(encoded_sequence)  
```

```python
"""
Encoiding
"""
text = "This is an example sentence."

# use .encode to encode a text
encoded_sequence = tokenizer.encode(text, add_special_tokens=True)
print(encoded_sequence)  
```

```python
"""
Decode 
"""
print(tokenizer.decode(encoded_sequence))
```
